```python
import h5py

# 확인하고 싶은 파일 경로를 입력하세요
file_path = "/home/najo/NAS/VLA/dataset/New_dataset/collected_data/Eye_trocar/Eye_trocar/260112/episode_20260112_151713.h5"  # 예: collected_data_sim 폴더 안의 파일

def print_structure(name, obj):
    # 데이터셋(실제 데이터)인 경우
    if isinstance(obj, h5py.Dataset):
        print(f"[Dataset] {name} \t Shape: {obj.shape} \t Type: {obj.dtype}")
    # 그룹(폴더 역할)인 경우
    elif isinstance(obj, h5py.Group):
        print(f"[Group]   {name}")

try:
    with h5py.File(file_path, 'r') as f:
        print(f"--- File: {file_path} Structure ---")
        f.visititems(print_structure) # 재귀적으로 모든 내부 구조를 탐색
except FileNotFoundError:
    print("파일을 찾을 수 없습니다. 경로를 확인해주세요.")
except Exception as e:
    print(f"오류 발생: {e}")
```
